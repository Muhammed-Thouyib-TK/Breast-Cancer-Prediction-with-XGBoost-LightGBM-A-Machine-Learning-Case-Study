```python
# importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset

df = pd.read_csv('breast cancer data.csv')
df
```

```python
# to get the basic info

df.info()
```

```python
# to get the basic stats of dataset

df.describe()
```

```python
# to check for null value

df.isnull().sum()
```

```python
# to remove unwanted columns

df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
df
```

```python
# to plot the target value

sns.countplot(df['diagnosis'])
plt.show()
```

```python
# mapping

df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
```

```python
df
```

```python
# to display all columns

pd.set_option('display.max_columns',None)
df.describe()
```

```python
# to get advanced stats of dataset

import Dis_Stats as ds

for i in df.columns:
    print('||',i,'||')
    ds.calc_fo(df[i])
    print('='*50)
```

```python
# to create some plots to analyze the dataset

for i in df.drop('diagnosis',axis=1).columns:
    print('||',i,'||')
    sns.displot(data=df,x=i,hue='diagnosis')
    plt.show()
```

```python
for i in df.drop('diagnosis',axis=1).columns:
    print('||',i,'||')
    sns.violinplot(data=df,x=i,hue='diagnosis')
    plt.show()
```

```python
for i in df.drop('diagnosis',axis=1).columns:
    print('||',i,'||')
    sns.kdeplot(data=df,x=i,hue='diagnosis')
    plt.show()
```

```python
# to split the dataset

x = df.drop('diagnosis',axis=1)
y = df['diagnosis']
```

```python
x
```

```python
y
```

```python
# to split the dataset into training and testing

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# XGBM & LGBM

import xgboost as xg
import lightgbm as lgm
```

```python
xgbm_model = xg.XGBClassifier()
lgbm_model = lgm.LGBMClassifier()
```

```python
# to fit the model

xgbm_model.fit(x_train,y_train)
```

```python
lgbm_model.fit(x_train,y_train)
```

```python
# to predict 

xgb_y_pred_train = xgbm_model.predict(x_train)
xgb_y_pred_test = xgbm_model.predict(x_test)
```

```python
xgb_y_pred_train
```

```python
xgb_y_pred_test
```

```python
lgm_y_pred_train2 = lgbm_model.predict(x_train)
lgm_y_pred_test2 = lgbm_model.predict(x_test)
```

```python
lgm_y_pred_train2
```

```python
lgm_y_pred_test2
```

```python
# to get the classification report

from sklearn.metrics import classification_report

print(classification_report(y_train,xgb_y_pred_train))
```

```python
print(classification_report(y_test,xgb_y_pred_test))
```
